### Import all Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Static Website name

```python
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D&list_price_max&maximum_distance=all&zip'
```

### Get Request, 200 --> Okay

```python
response = requests.get(website)
response.status_code
```

### Define Soup

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

## Single Page Web Scraping

### Define Lists

```python
name=[]
mileage=[]
dealer_name = []
rating = []
rating_count = []
price = []
```

### Loop Through the WebPage and append Data

```python
for result in results:
    name.append(result.find('h2').get_text())
    mileage.append(result.find('div', {'class':'mileage'}).get_text())
    dealer_name.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
    try:
        rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
    except:
        rating.append('n/a')
    try:
        rating_count.append(result.find('span', {'class':'sds-rating__link'}).get_text())
    except:
        rating_count.append('n/a')
    price.append(result.find('span', {'class':'primary-price'}).get_text())
```

### Create Database

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Dealer Name': dealer_name, 'Rating':rating, 'Rating Count':rating_count, 'Price':price})
```

```python
car_dealer
```

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
```

```python
car_dealer
```

### Save as Excel

```python
car_dealer.to_excel('single_page_car.xlsx', index=False)
```

## Pagination using Dynamic Website Name

### Defining Lists

```python
name=[]
mileage=[]
dealer_name = []
rating = []
rating_count = []
price = []
```

### Appending Data on all Pages

```python
for i in range(1,11):
    website = 'https://www.cars.com/shopping/results/?list_price_max=&makes[]=mercedes_benz&maximum_distance=all&models[]=&page=' + str(i) + '&stock_type=cpo&zip='
    response = requests.get(website)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all('div', {'class':'vehicle-card'})
    for result in results:
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')
        try:
            mileage.append(result.find('div', {'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')
        try:
            dealer_name.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')
        try:
            rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')
        try:
            rating_count.append(result.find('span', {'class':'sds-rating__link'}).get_text())
        except:
            rating_count.append('n/a')
        try:
            price.append(result.find('span', {'class':'primary-price'}).get_text())
        except:
            price.append('n/a')    
```

### Create DataFrame

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Dealer Name': dealer_name, 'Rating':rating, 'Rating Count':rating_count, 'Price':price})
```

```python
car_dealer
```

```python
car_dealer['Rating Count'] = car_dealer['Rating Count'].apply(lambda x:x.strip('reviews)').strip('('))
```

### Save as Excel

```python
car_dealer.to_excel('multi_page_car.xlsx', index=False)
```
